```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from PIL import Image
import os
import numpy as np
```

```python
base_dir = "train_set" 
images_dir = base_dir+"/images"
expressions = ["Neutral", "Happy", "Sad", "Surprise", "Fear", "Disgust", "Anger", "Contempt"]
for i, name in enumerate(os.listdir(images_dir)):
    print(name)
    if i == 5:
        break
    number = name[0:-4]
    img = Image.open(images_dir +"/"+ name).convert("L")
    print(np.array(img).shape)
    imgplot = plt.imshow(img, cmap="gray")
    plt.show()
    
    print(np.load(f"{base_dir}/annotations/{number}_aro.npy"), np.load(f"{base_dir}/annotations/{number}_val.npy"))
    # print("lnd:", np.strings.decode(np.load(f"{base_dir}/annotations/{number}_lnd.npy"), encoding="utf-8"))
    exp_index = int(np.load(f"{base_dir}/annotations/{number}_exp.npy"))
    print("exp:", expressions[int(exp_index)])
    
```
